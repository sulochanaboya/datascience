```python
data=pd.read_csv("student_data.csv")
print(data)
```

```python
plt.scatter(data['Hours'],data['Marks'])
plt.xlabel("study Hours")
plt.ylabel("Marks")
plt.title("Hours vs Marks")
plt.show()
```

```python
x=data[['Hours']]
y=data['Marks']
```
