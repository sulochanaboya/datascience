```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

categories = ['Technology', 'Furniture', 'Office Supplies']
regions = ['East', 'West', 'Central', 'South']
segments = ['Consumer', 'Corporate', 'Home Office']

sales_data = {
    'Order Date': pd.date_range(start='2025-01-01', periods=n, freq='D'),
    'Category': np.random.choice(categories, n),
    'Region': np.random.choice(regions, n),
    'Segment': np.random.choice(segments, n),
    'Quantity': np.random.randint(1, 10, n),
    'Sales': np.random.randint(100, 5000, n),
    'Profit': np.random.randint(10, 1000, n)
}

df = pd.DataFrame(sales_data)

df.to_csv('retail_sales.csv', index=False)

print('CSV file created successfully!')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("retail_sales.csv")
print(df.head())
df.isnull().sum()
df['Order Date']=pd.to_datetime(df['Order Date'])
monthly_sales=df.groupby(df['Order Date'].dt.month)['Sales'].sum()
monthly_sales.plot(kind='line')
plt.title("monthly sales Trend")
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
sns.barplot(x='category',y='sales',data=df)
plt.title("sales by category")
plt.show()

region_sales=df.groupby('Region')['sales'].sum()
region_sales.plot(kind='pie',autopct='%1.1f%%')
plt.show()
#model training
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x=df[['quantity']]
y=df['sales']

x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
model=LinearRegression()
model.fit(x_train,y_train)

predictions=model.predict(x_test)

from sklearn.metrics import mean_absolute_error,r2_score
print("MAE:",mean_absolute_error(y_test,predictions))
print("R2 score:",r2_score(y_test,predictions))
```
